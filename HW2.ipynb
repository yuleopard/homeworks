3. Пройтись циклом по рабочей директории, читая содержимое каждого файла и записывая его в новый файл poems.txt По завершении цикла должен получиться файл со всеми стихотворениями!

```python
import os
```

```python
# Распечатать текущую директорию и список всех файлов и папок в ней. Посчитать и распечатать количество файлов.

# делаем список, в который будут добавляться имена файлов
allfiles = [] 

# цикл для получения файлов
for root, dirs, files in os.walk('.'):
    for name in files:
        
        # составляем полный путь к файлу и печатаем
        fullname = os.path.join(root, name)
        print(fullname)
        
    # добаляем имя в список, чтобы посчитать количество
    allfiles.append(files)
    
print('Количество файлов: ', len(allfiles))
```

```python
# Пройтись циклом по рабочей директории, читая содержимое каждого файла и записывая его в новый файл poems.txt 
for root, dirs, files in os.walk('./poems'):
    for file in files:
        
        # открываем файл из папки
        with open(os.path.join(root, file), 'r', encoding='utf-8') as f:
            poem = f.read()
        
        # добавляем только что открытое стихотворение в общий файл
        with open('./poems/poems.txt', 'a', encoding='utf-8') as f:
            f.write(poem + '\n')
```

```python
#1. Скопировать текст, сохранить его в переменную.  
#2. Узнать длину текста в символах.  

poem = '''Несчастная кошка порезала лапу- 
Сидит, и ни шагу не может ступить.
Скорей, чтобы вылечить кошкину лапу
Воздушные шарики надо купить!

И сразу столпился народ на дороге-
Шумит, и кричит, и на кошку глядит.
А кошка отчасти идет по дороге,
Отчасти по воздуху плавно летит!'''
print('Количество символов: ', len(poem))
```

```python
#3. Очистить текст от пунктуации, используя list compehensions. На выходе должен получиться список слов.  
poemclean = [w.strip("!?.,-") for w in poem.split()]
print(poemclean)
```

```python
#4. Объединить слова из получившегося списка в строку через пробел.  
poemstring = ' '.join(poemclean)
print(poemstring)
```

```python
#5. С помощью list comprehensions проверить, написано ли слово с большой буквы, и если да, то привести его к нижнему регистру. 
# На выходе должен получиться список из всех слов стихотворения, написанных с маленькой буквы 
#(не только те, которые изначально были с маленькой буквы, а все). 
poemlower = [w if w.islower() else w.lower()  for w in poemstring.split()]
print(poemlower)
```

```python
#6. Узнать длину текста в словах.  
print('Количество слов: ', len(poemlower))
```

```python
#7. Узнать количество уникальных слов. 
# делаем из списка массив, чтобы элементы не повторялись
poemset = set(poemlower)
print('Количество уникальных слов :', len(poemset))
```
